```python
n = 1000
k = 400
lenght = 219

avg_life_exp = []

for i in range(lenght):
    db = np.random.randint(0, 6, int(child_mort.mean()[i]))
    ndb = np.random.binomial(80,0.65, 1000 - int(child_mort.mean()[i]))
    ppl_1000_life_exp = np.append(db, ndb)
    avg_life_exp.append(np.mean(ppl_1000_life_exp))
    
```

```python
plt.plot(avg_life_exp)
life_exp.mean().plot()
plt.plot([128,128 ], [20, 70])
```

```python
max(avg_life_exp)
```

```python
avg_life_exp_rolling_avg_7 = []
for i in range(0,len(avg_life_exp), 7):
    avg_life_exp_rolling_avg_7.append(sum(avg_life_exp[i:i+7])/7)
    
    
    
```

```python
plt.plot(avg_life_exp_rolling_avg_7[:-2])
```

```python
derivates = [avg_life_exp[x+1] - avg_life_exp[x] for x in range(len(avg_life_exp)-1)]
```

```python
der_list = []
for i in range(0,len(derivates), 7):
    der_list.append(sum(derivates[i:i+7])/7)
    
    
```

```python
plt.plot(der_list)
```
